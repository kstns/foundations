## Assignment

- Import pandas and matplotlib (remember "% matplotlib inline")
- Read in the data and explore the head and tail
- Drop any rows with missing values

```python
import pandas as pd
import matplotlib
% matplotlib inline

dir = r"C:\Users\Kostya\K2\foundations\college_majors\college_majors\recent-grads.csv"
college = pd.read_csv(dir)
print(college.head())
print(college.tail())
college.dropna(inplace = True)
```

## Assignment

Feel free to use native panda graphing methods.

- Make a scatter plot of sample size vs. median
- Make a scatter plot of sample size vs. unemployment rate
- Make a scatter plot of full time vs. median
- Make a scatter plot of men vs. median
- Make a scatter plot of women vs. median


```python
college.plot(kind="scatter", x="Sample_size", y="Median")
college.plot(kind="scatter", x="Sample_size", y="Unemployment_rate")
college.plot(kind="scatter", x="Full_time", y="Median")
college.plot(kind="scatter", x="Men", y="Median")
college.plot(kind="scatter", x="Women", y="Median")
```
